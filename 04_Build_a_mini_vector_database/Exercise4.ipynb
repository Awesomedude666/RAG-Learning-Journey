```python
# Exercise 4 : Build a mini vector db
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer("all-MiniLM-L6-v2")
vector_db = []
```

```python
def add_document(text):
    doc_embedding = model.encode(text)

    vector_db.append({
        "text":text,
        "embedding":doc_embedding
    })
```

```python
def search(query):
    query_embedding = model.encode([query])

    best_score = -1
    best_doc = None

    for doc in vector_db:
        score = cosine_similarity(
            query_embedding,
            [doc["embedding"]]
        )[0][0]

        if score > best_score:
            best_score = score
            best_doc = doc["text"]

    return best_doc, best_score
 
```

```python
# Add documents
add_document("I love machine learning")
add_document("I enjoy deep learning")
add_document("Football is a sport")

# Search
doc, score = search("machine learning")

print("Best Match:")
print(doc)

print("\nSimilarity Score:")
print(score)
```
